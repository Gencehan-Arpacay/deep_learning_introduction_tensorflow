## Steps with modeling with tensorflow

1. Creating a model- Define the input and the output layers including the hidden layers of deep learning model

2. Compling a model - define the loss function in other words tell the model who wrong it is and the optimizer tells the model how to improve the model and evaluation matrix

3. Fitting a model - letting the model try to find patterns between x and y

```python
# Turn np arrays to tensors

X = tf.cast(tf.constant(X), dtype = tf.float32)

y = tf.cast(tf.constant(y), dtype = tf.float32)


X, y
```

```python
y == X + 10
```

```python
inp_shape = X[0].shape
out_shape = y[0].shape

inp_shape, out_shape
```

```python
# set random seed
tf.random.set_seed(42)

# 1. Creating a model using sequantial API

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)), # this has resolved the issue of the dimensions error https://stackoverflow.com/questions/72423651/valueerror-for-sequential-model-fit
  tf.keras.layers.Dense(1)
])


# 2. Compile the model

model.compile(loss=tf.keras.losses.mae, #mea stands for mean absolute error
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

# 3. Fit the model
model.fit(X, y, epochs = 5)
```

```python
model.predict([17.0])
```

```python
X_train = X[:40]
y_train = y[:40]

X_test = X[40:]
y_test = y[40:]

len(X_train), len(X_test)
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, input_shape=[1], name="input_layer"),
    tf.keras.layers.Dense(1, name="output_layer")
])

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.1),
              metrics=["mae"],
              )
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 100, verbose = 0)
```

```python
# Lets create a plotting function

def plot_predictions(train_data=X_train, train_labels=y_train, test_data = X_test, test_labels = y_test, predictions=y_pred):
  """
  Plots training and test data and compers predictions with real results
  """
  plt.figure(figsize=(10,7))
  plt.scatter(train_data, train_labels, c="b", label="Training data")
  plt.scatter(test_data, test_labels, c="g", label="Testing data")
  plt.scatter(test_data, predictions, c="r", label="Predictions")
  plt.legend()
```

```python
plot_predictions(train_data=X_train,
                 train_labels=y_train,
                 test_data = X_test,
                 test_labels = y_test,
                 predictions=y_pred)
```

### Evaluating our model's predictions with regression evaluation metrics


depending on the problem we are working on there are different evaluation metrics since we are working on regression problems two main metrics:
* MAE - mean absolute error, "on avg how wrong each models pred"
> * ${MAE} = (\sum \limits _{i=1} ^{n}  y_{i}x_{i}) / n $
> * tf.keras.losses.MAE
> * It is a greate metric for any regression problem
* MSE mean squared error
> * ${MSE} = 1/n \sum \limits _{i=1} ^{n}  (Y_{i} - \hat{Y_{i}})^2 $
> * Hat y is the y predictions 
> * tf.keras.losses.MSE
> * We use this when larger errors are more significant 
* Huber:
> * tf.keras.Huber()
> * Combination of MSE and MAE less sensitive to outliers than MSE



```python
#Â Evaluate the model on the test:

model.evaluate(X_test, y_test)
```
